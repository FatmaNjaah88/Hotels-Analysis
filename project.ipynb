Data Analysis & visualizations

```python
canceled_percentage=data['is_canceled'].value_counts(normalize=True)
canceled_percentage
```

```python
canceled_percentage=data['is_canceled'].value_counts(normalize=True)
print(canceled_percentage)
plt.figure(figsize=(5,4))
plt.title("Reservation status count")
plt.bar(["Not canceled","canceled"],data['is_canceled'].value_counts(),edgecolor='k',width=.7)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='hotel',hue='is_canceled',data=data,palette='Pastel1')
plt.title("Calculation rates in city hotel and resort hotel",fontweight='bold',size=18)
plt.show()
```

Most booking were in city hotel
Cancelation in resort hotel is less than city hotel

```python
city_hotel=data[data['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=data[data['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize=(20,8),facecolor='#D4E6F1')
plt.title("Average daily rate in city  and resort hotel ",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
data['month']=data['reservation_status_date'].dt.month
plt.figure(figsize=(15,7),facecolor='#D4E6F1')
sns.countplot(x='month',hue='is_canceled',data=data,palette='Pastel1')
plt.title("Reservation status  per month",fontweight='bold',size=20)
plt.xlabel('month',fontweight='bold',size=15)
plt.ylabel('numder of reservations',fontweight='bold',size=15)
plt.legend(['not canceled','canceled'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',size=30)
df=data[data['is_canceled']==1].groupby('month')['adr'].sum().reset_index()
colors = sns.color_palette("husl", len(df))
sns.barplot(x='month',y='adr',data=df,palette=colors)
plt.show()
```

```python
canceled_data=data[data['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with reservation canceled",color='black')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
data['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date',as_index=False)['adr'].mean()
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=data[data['is_canceled']==0]
not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date',as_index=False)['adr'].mean()
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)
plt.figure(figsize=(20,6),facecolor='#D4E6F1')
plt.title("Average daily rate",size=40,color='black')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend(fontsize=20)
plt.show()

```

```python
canceled_df_adr['reservation_status_date'].info()
```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016-01-01') & 
                                  (canceled_df_adr['reservation_status_date'] < '2017-09-01')]

not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016-01-01') & 
                                          (not_canceled_df_adr['reservation_status_date'] < '2017-09-01')]
```

```python
plt.figure(figsize=(20,6),facecolor='#D4E6F1')
plt.title("Average daily rate",size=40,color='black')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend(fontsize=20)
plt.show()
```
